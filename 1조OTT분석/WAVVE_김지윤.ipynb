### 1) 연령별_20세미만

```python
# 모듈 로딩
import pandas as pd
import numpy as np

# 파일로딩
DIR='../Data/'
FILE=DIR+'웨이브_연령_20세미만.xlsx'
teen=pd.read_excel(FILE, skiprows=6)
teen
```

```python
# NaN 값 없애기
teen=teen.dropna()
teen.isnull().sum()
```

```python
# 날짜 타입 변환
teen['날짜']=teen['날짜'].astype('datetime64')
teen
```

```python
# month 열 지정
M=teen['날짜'].dt.month
teen.insert(1, "MONTH", M)
teen
```

```python
# 날짜 열 인덱스 지정
teen.set_index('날짜', inplace =True)
```

```python
# 2019년에 해당하는 행 추출
Y2019=teen.loc['2019-01-01':'2019-12-31']
Y2019   
```

```python
# 2019년 월별 평균 검색 수
monthGroup_2019=Y2019.groupby(by='MONTH')
mean_2019=monthGroup_2019.mean()
mean_2019
```

```python
# 2020년에 해당하는 행 추출
Y2020=teen.loc['2020-01-01':'2020-12-31']

# 2019년 월별 평균 검색 수
monthGroup_2020=Y2020.groupby(by='MONTH')
mean_2020=monthGroup_2020.mean()
mean_2020
```

```python
# 2021년에 해당하는 행 추출
Y2021=teen.loc['2021-01-01':'2021-12-31']

# 2021년 월별 평균 검색 수
monthGroup_2021=Y2021.groupby(by='MONTH')
mean_2021=monthGroup_2021.mean()
mean_2021
```

```python
# 2022년에 해당하는 행 추출
Y2022=teen.loc['2022-01-01':'2022-06-30']

# 2022년 월별 평균 검색 수
monthGroup_2022=Y2022.groupby(by='MONTH')
mean_2022=monthGroup_2022.mean()
mean_2022
```

```python
wavve_teen=pd.concat([mean_2019, mean_2020, mean_2021,mean_2022], ignore_index=True)
wavve_teen
```

### 2) 연령별 _ 20대

```python
# 파일로딩
DIR='../Data/'
FILE=DIR+'웨이브_연령_20대.xlsx'
twenty=pd.read_excel(FILE, skiprows=6)
twenty.tail()
```

```python
# 반복되는 코드 함수 만들기
# NaN값 없애는 함수
def dropNaN(df_name):
    df_name=df_name.dropna()
    

# 날짜열 타입 바꾸는 함수    
def changeDateType(df_name):
    df_name['날짜']=df_name['날짜'].astype('datetime64')
    
    
# month열 생성하고 인덱스 지정하는 함수
def monthColumn(df_name):
    M=df_name['날짜'].dt.month
    df_name.insert(1, 'MONTH', M)
    df_name.set_index('날짜', inplace =True)
    
    
# 각 연도 별, 월 별 검색 건수 평균 구하는 함수
def yearMean(df_name):
    global mean_2019, mean_2020, mean_2021, mean_2022
    Y2019=df_name.loc['2019-01-01':'2019-12-31']
    monthGroup_2019=Y2019.groupby(by='MONTH')
    mean_2019=monthGroup_2019.mean()
    
    Y2020=df_name.loc['2020-01-01':'2020-12-31']
    monthGroup_2020=Y2020.groupby(by='MONTH')
    mean_2020=monthGroup_2020.mean()
    
    Y2021=df_name.loc['2021-01-01':'2021-12-31']
    monthGroup_2021=Y2021.groupby(by='MONTH')
    mean_2021=monthGroup_2021.mean()
    
    Y2022=df_name.loc['2022-01-01':'2022-06-30']
    monthGroup_2022=Y2022.groupby(by='MONTH')
    mean_2022=monthGroup_2022.mean()
```

```python
# NaN값 없애기
dropNaN(twenty)

# 날짜열 타입 변환하기
changeDateType(twenty)

# MONTH열 생성 후 인덱스로 지정하기
monthColumn(twenty)

# 각 연도별, 월별 검색 건수 평균 구하기
yearMean(twenty)

# 검색 건수 평균 전체 연도 합치기
wavve_twenty=pd.concat([mean_2019, mean_2020, mean_2021, mean_2022], ignore_index=True)

# 데이터 확인
wavve_twenty
```
