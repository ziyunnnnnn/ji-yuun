(항목 varchar(10), 시점 int, 대한민국 varchar(10), 일본 varchar(10), 이탈리아 varchar(10), 포르투갈 varchar(10));

insert 노령화지수(항목, 시점, 대한민국, 일본, 이탈리아, 포르투갈)

create table 노년총부양비

insert 노년총부양비(항목, 시점, 대한민국, 일본, 이탈리아, 포르투갈)

create table 경제상태만족도_연령별

create table 건강상태만족도_연령별
as(

from 건강상태만족도

create table 노인_자살율
(특성별 varchar(50), 특성 varchar(3), 전국 int);

insert 노인_자살율
select 특성별, 특성, 전국
from 노인자살율
where 특성 = '계';

select * from 노인_자살율;

```python
# 모듈로딩
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import platform
import seaborn as sns
```

```python
# sql, 파이썬 연동 함수
def sql(table_name):
    import pymysql
    conn = pymysql.connect(host='localhost', user='root', password='691209', db = 'project', charset='utf8')
    cur = conn.cursor()
    query = """
    select * from {};
    """
    cur.execute(query.format(table_name))
    rows = cur.fetchall() 
    df=pd.DataFrame(rows)
    cur.close()
    conn.close()
    return df
```

```python
df=sql('노령화지수')
df.columns=['항목', '시점', '대한민국', '일본', '이탈리아', '포르투갈']
df.replace('-', 576.6, inplace=True)
df=df.astype({'대한민국':'float','일본':'float', '이탈리아':'float', '포르투갈':'float'})
```

```python
colors = sns.color_palette("Set3", 10)
w=0.215
nrow = df.shape[0] 
idx = np.arange(nrow)

plt.figure(figsize = (18, 8))
plt.title('노령화지수', fontsize=20)
plt.xlabel('년도')
plt.ylabel('노령화지수')
plt.bar(idx - 2 * w, df['대한민국'], width = w, color=colors[0],label = '대한민국')
plt.plot(idx - 2 * w, df['대한민국'], 'o--', markersize=17, color='slategray', linewidth=5, alpha=0.85)
plt.bar(idx - w, df['일본'], width = w, color=colors[1], label = '일본')
plt.bar(idx, df['이탈리아'], width = w, color=colors[2],label = '이탈리아')
plt.bar(idx + w, df['포르투갈'], width = w, color=colors[3],label = '포르투갈')
plt.xticks(idx, df['시점'], rotation = 30)
plt.legend(ncol = 5)

plt.style.use('ggplot')
plt.show()
```
