```python
# import_ipynb 임포트
import import_ipynb
# import_pandas 임포트
import pandas as pd
# seaborn 임포트
import seaborn as sns
# matplotlib.pyplot 임포트
import matplotlib.pyplot as plt
```

```python
# 각각의 분석 파일 import
import 인구밀도_이재훈 as pop
import 이혼율_심정윤 as divorce
import 실업율_임재원 as unemploy
import 물가지수_부동산지수_이호진 as money
import 결혼율_교육비부담도_김지윤 as wedding
```

```python
# 인구밀도 파일에서 해당 연도에 따른 인구밀도 리스트 추출
corr_pop = pop.ration_list
# 실업율 파일에서 해당 연도에 따른 실업율 리스트 추출
corr_unemploy = unemploy.age_list
# 물가 파일에서 해당 연도에 따른 물가 및 부동산 가격 리스트 추출
corr_money = money.corr_money
corr_build = money.corr_build
# 이혼율 파일에서 해당 연도에 따른 이혼율 리스트 추출
corr_DivRate = divorce.corr_DivRate
# 결혼율 파일에서 해당 연도에 따른 결혼건수 리스트 추출
corr_wedding = wedding.corr_wedding
```

```python
# 변수들 간의 데이터를 데이터 프레임
df = pd.DataFrame({'인구밀도': corr_pop, '실업율' : corr_unemploy, '물가지수' : corr_money, '부동산지수' : corr_build, '이혼율' : corr_DivRate, '결혼건수' : corr_wedding})
```

```python
df
```

```python
# 각 변수 간 상관계수 계산
corr_df = df.corr()
corr_df
```

```python
# 상관계수 데이터프레임 시각화
plt.figure(figsize=(13, 13)) # 시각화 자료 크기 설정
sns.heatmap(corr_df, annot=True, cmap='Reds', linewidths= 2, linecolor='white', center = 0.5) # 시각화 자료 크기 설정
plt.show()
```
