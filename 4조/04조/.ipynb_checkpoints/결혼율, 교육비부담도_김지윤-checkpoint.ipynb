# 자료 로딩

```python
# -- 출산율 --
# 모듈 로딩
import pandas as pd 

# 파일관련 변수 선언 
DIR_PATH='./Data/'
FILE_NAME=DIR_PATH+'출산율.xlsx'

# DF 형태 데이터로 로딩
birth=pd.read_excel(FILE_NAME, skiprows=2)
birth.info()
```

```python
# -- 결혼건수 --
# 파일관련 변수 선언
DIR_PATH='./Data/'
FILE_NAME=DIR_PATH+'결혼율.csv'

# DF 형태 데이터로 로딩
marriage=pd.read_csv(FILE_NAME)
marriage.info()
```

```python
# -- 교육비 부담도 --
# 파일관련 변수 선언
DIR_PATH='./Data/'
FILE_NAME=DIR_PATH+'교육비 부담도.xls'

# DF 형태 데이터로 로딩
edu=pd.read_excel(FILE_NAME, skiprows=2)
edu.info()
```

# 데이터 전처리

```python
# -- 출산율 --
# 데이터 확인
birth
```

```python
# 2,3행 삭제
birth=birth.drop([2,3])
birth
```

```python
# 2011~2020 열 추출
birth=birth.iloc[:,-10:]
```

```python
# 인덱스 라벨 바꾸기
birth=birth.rename(index={0:'출생아수', 1:'합계출산율'})
birth
```

```python
# -- 결혼건수 --
# 데이터 확인
marriage
```

```python
# 전국, 서울, 대구 행만 불러오기
marriage=marriage.iloc[[0,1,3]]
```

```python
# 2011~2020의 6월 자료만 추출
year_list=[]
for i in range(2011, 2021):
    data=i
    data=str(data)
    data2=data+'.06'
    year_list.append(data2)
year_list
marriage=marriage.loc[:,year_list]
```

```python
# 인덱스 설정
marriage.index=['전국', '서울', '대구']
# 컬럼명 설정
marriage=marriage.rename(columns={'2011.06':'2011', '2012.06':'2012','2013.06':'2013','2014.06':'2014','2015.06':'2015','2016.06':'2016','2017.06':'2017','2018.06':'2018','2019.06':'2019','2020.06':'2020'})
marriage
```

```python
# -- 교육비 부담도 --
# 데이터확인
edu
```

```python
# 2008, 2010 열 삭제
edu=edu.drop(['2008', '2010'], axis=1)

# 출처, 주석 행 삭제
edu=edu.drop([7,8,9])
edu
```

# 상관계수 

```python
# -- 결혼건수/출산율 상관계수

# 전국 결혼 건수만 추출
m=marriage.loc['전국']

#출생아수만 추출
b=birth.iloc[0]

# 타입 통일
b1=b.astype(int)
m1=m.astype(int)

print(b1, m1)
```

```python
# 시리즈 두 개 합치기
result1=pd.concat([b1,m1],axis=1)

# 컬럼명 바꾸기 
result1=result1.rename(columns={'전국':'결혼건수'})
result1
```

```python
# 결혼건수와 출생아수의 상관계수
result1.corr()
```

```python
# -- 교육비부담도/결혼건수 상관계수
# 교육비 부담도 df에서 0번 행과 짝수년도만 추출
edu1=edu.loc[0, ['2012', '2014', '2016', '2018', '2020']]
edu1=edu1.astype(float)

# 결혼율 df에서 '전국'열과 짝수년도 열만 추출
m=marriage.loc['전국', ['2012', '2014', '2016', '2018', '2020']]
m=m.astype(float)

# 교육비 부담도와 결혼율에서 추출한 시리즈 합치기
result2=pd.concat([edu1,m],axis=1)

# 컬럼명 바꾸기
result2=result2.rename(columns={0:'교육비 부담도', '전국':'결혼건수'})

result2
```

```python
# 교육비 부담도와 결혼건수의 상관계수
result2.corr()
```

```python
# -- 교육비부담도/출산율 상관계수
# 교육비 부담도 시리즈, 출산아수 시리즈 합치기
result3=pd.concat([edu1,b1],axis=1)

# 연도 통일
result3=result3.drop(['2011','2013', '2015', '2017', '2019'])

# 컬럼명 바꾸기
result3=result3.rename(columns={0:'교육비 부담도'})

result3
```

```python
# 결혼 건수, 출생아수 상관계수
result3.corr()
```

# 그래프

```python
import matplotlib.pyplot as plt
```

```python
# matplotlib 한글 font 설정
from matplotlib import font_manager as fm, rc

# 한글폰트 설정
font_path='./Fonts/malgun.ttf'
font_name=fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
# -- 결혼건수/출산율 비교 그래프 --
```

```python
result1
```

```python
# x축, y축 지정
x=result1.index
y1=result1['출생아수'].to_list()
y2=result1['결혼건수'].to_list()

# 다중 플롯 지정
fig, ax1 = plt.subplots()
ax1.plot(x, y1, label='출산아수', color = 'red')
ax1.set_ylabel('출산아수', color = 'red')
ax1.legend()

# x축 공유
ax2 = ax1.twinx()
ax2.plot(x, y2, label='결혼건수 ',color = 'blue')
ax2.set_ylabel('결혼건수', color = 'blue')

ax2.legend(loc='upper left')

plt.title('결혼건수-출산율')
ax1.set_xlabel('년도')
plt.show()
```

```python
# -- 교육비부담도-결혼건수 그래프 --
```

```python
result2
```

```python
x=result2.index
y1=result2['교육비 부담도'].to_list()
y2=result2['결혼건수'].to_list()

# ax1은 y1에 대한 그래프
fig, ax1 = plt.subplots()
ax1.plot(x, y1, label='교육비 부담도', color = 'black')
ax1.set_ylabel('교육비 부담도', color = 'black')
ax1.legend()

# twinx로 x축을 공유
ax2 = ax1.twinx()
ax2.plot(x, y2, label='결혼건수', color = 'purple')
ax2.set_ylabel('결혼건수', color = 'purple')
ax2.legend(loc='upper left')

plt.title('교육비 부담도-결혼건수')
ax1.set_xlabel('년도')
plt.show()
```
