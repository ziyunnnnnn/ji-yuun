# 저출산과 실업율의 상관관계에 대해서 분석해본다.

```python
# 모듈 삽입
import pandas as pd

# 파일경로 관련 변수 선언
DIR_PATH = 'Data/'
FILE_AGE = '연령별_경제활동인구.csv'
FILE_AGE_GENDER = '성,연령별_경제활동인구.csv'
FILE_BIRTH = '출산율(재원).csv'
```

```python
# 출산율 데이터 프레임 객체 생성
birthDF = pd.read_csv(DIR_PATH + FILE_BIRTH, encoding='cp949', header=2, index_col = 0)    # 불필요한 헤더 제거 및 인덱스 제거
birthDF
```

```python
# 출산율 데이터프레임 가공

# 불필요한 행 drop
birthDF.drop(['출생아?수','출처:', '주석:'], inplace = True)    # 쓰지 않을 행이기 때문에 inplace = True

# 행 인덱스 이름 변경
birthDF.rename(index={'합계?출산율':'출산율'}, inplace=True)

# 필요한 년도만 추출(2011 ~ 2020년)
birthDF = birthDF.iloc[:, 41:]

# 정제된 출산율 데이터 확인
birthDF
```

```python
# 전체 연령별 고용현황 데이터 프레임 객체 생성
ageDF = pd.read_csv(DIR_PATH + FILE_AGE, encoding='cp949', header=0)
ageDF
```

```python
# 전체 연령별 고용현황 데이터 프레임 가공

# 결혼 적령기인 20~39세 데이터 추출 및 여러 데이터중 실업율에 해당하는 부분만 추출
year_list = ['연령계층별']
for i in range(2011,2021):
    data = i
    data = str(data)
    data_2 = data + '.6'
    year_list.append(data_2)

# 인덱스 '연령계층별'로 설정
ageDF = ageDF.loc[3:4, year_list].set_index('연령계층별')
ageDF
```

```python
# 정제된 데이터로의 계산에 앞서 결측치가 없는지 확인
print(ageDF.isnull().sum())

# 결측치가 관찰되지 않으므로 결측치에 대한 처리는 생략함
```

```python
# 20~29세, 30~39세 실업율 평균 구한 후 하나의 row로 합치기

# 평균을 내기에 앞서 정제한 데이터의 type을 확인
print(ageDF.info())

# type이 object로 되어있음을 확인, float 형태로 변환
ageDF = ageDF.astype(float)

# 다시 type을 확인, float 형태로 바뀌었음을 확인
print(ageDF.info())
```

```python
# decribe 메서드를 사용, 평균을 구한 뒤 새로운 dataframe 생성
print(ageDF.describe())
print(ageDF.describe().loc['mean'])
year = []
for i in range(2011,2021):
    year.append(i)
ageDF=pd.DataFrame(ageDF.describe().loc['mean']).transpose()

# row index 변경
ageDF.rename({'mean':'결혼 적령기 실업율(%)'}, inplace=True)

# 데이터 프레임을 합치기 위한 column index 통일화 작업
ageDF.columns = birthDF.columns.to_list()

# 정제된 연령별 실업율 data 확인
ageDF
```

```python
# 출산율 data와 실업율 data 합치기
finalDF = pd.concat([birthDF,ageDF])
finalDF
```

```python
# 같은 방식으로 성별, 연령별 고용현황 데이터 정제
age_genderDF = pd.read_csv(DIR_PATH + FILE_AGE_GENDER, encoding='cp949', header=0)
age_genderDF
```

```python
# 결혼 적령기인 20~39세의 남자 데이터 추출 및 여러 데이터중 실업율에 해당하는 부분만 추출
year_list = ['성별', '연령계층별']
for i in range(2011,2021):
    data = i
    data = str(data)
    data_2 = data + '.6'
    year_list.append(data_2)
    
# 인덱스 '연령계층별'로 설정
age_maleDF = age_genderDF.loc[13:14, year_list].set_index(['연령계층별', '성별'])
age_maleDF
```

```python
# 20~29세, 30~39세 실업율 평균 구한 후 하나의 row로 합치기

# 평균을 내기에 앞서 정제한 데이터의 type을 확인
print(age_maleDF.info())

# type이 object로 되어있음을 확인, float 형태로 변환
age_maleDF = age_maleDF.astype(float)

# 다시 type을 확인, float 형태로 바뀌었음
print(age_maleDF.info())
```

```python
# decribe 메서드를 사용, 평균을 구한 뒤 새로운 dataframe 생성
print(age_maleDF.describe())
print(age_maleDF.describe().loc['mean'])
year = []
for i in range(2011,2021):
    year.append(i)
age_maleDF=pd.DataFrame(age_maleDF.describe().loc['mean']).transpose()

# row index 변경
age_maleDF.rename({'mean':'결혼 적령기 실업율(남자)(%)'}, inplace=True)

# 데이터 프레임을 합치기 위한 column index 통일화 작업
age_maleDF.columns = birthDF.columns.to_list()

# 정제된 남성 연령별 실업율 data 확인
age_maleDF
```

```python
birthDF.columns.to_list()
```

```python
# data 합치기
finalDF = pd.concat([finalDF,age_maleDF])
finalDF
```

```python
# 같은 방식으로 결혼 적령기인 20~39세의 여자 데이터 추출 및 여러 데이터중 실업율에 해당하는 부분만 추출
year_list = ['성별', '연령계층별']
for i in range(2011,2021):
    data = i
    data = str(data)
    data_2 = data + '.6'
    year_list.append(data_2)

# 인덱스 '연령계층별'로 설정
age_femaleDF = age_genderDF.loc[23:24, year_list].set_index(['연령계층별', '성별'])
age_femaleDF['2011.6'].to_list()
```

```python
# type이 object로 되어있음을 확인, float 형태로 변환
age_femaleDF = age_femaleDF.astype(float)
```

```python
# decribe 메서드를 사용, 평균을 구한 뒤 새로운 dataframe 생성
print(age_femaleDF.describe())
print(age_femaleDF.describe().loc['mean'])
year = []
for i in range(2011,2021):
    year.append(i)
age_femaleDF=pd.DataFrame(age_femaleDF.describe().loc['mean']).transpose()

# row index 변경
age_femaleDF.rename({'mean':'결혼 적령기 실업율(여자)(%)'}, inplace=True)

# 데이터 프레임을 합치기 위한 column index 통일화 작업
age_femaleDF.columns = birthDF.columns.to_list()

# 정제된 여성 연령별 실업율 data 확인
age_femaleDF
```

```python
# data 합치기
finalDF = pd.concat([finalDF,age_femaleDF])

# 최종 dataframe 확인
finalDF
```

```python
# 최종적으로 얻어진 data로 graph를 그려 데이터 시각화

# matplotlib 모듈 로딩
import matplotlib.pyplot as plt
```

```python
# 폰트 관련 설정
from matplotlib import font_manager as fm, rc

# 한글 폰트 설정
font_path = './Fonts/malgun.ttf'
font_name = fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
# 그래프에 사용하기 위해 dataframe에서 list 추출
birth_list = finalDF.loc['출산율'].to_list()
age_list = finalDF.loc['결혼 적령기 실업율(%)'].to_list()
male_list = finalDF.loc['결혼 적령기 실업율(남자)(%)'].to_list()
femali_list = finalDF.loc['결혼 적령기 실업율(여자)(%)'].to_list()
```

```python
# 그래프로 시각화
plt.plot(year, birth_list, 'mo-', label='출산율', ms = 4)
plt.legend()    # 범례
plt.plot(year, age_list, 'go--', label='실업율', ms = 4)
plt.legend()  
plt.plot(year, male_list, 'bo--', label='실업율(남)', ms = 4)
plt.legend()  
plt.plot(year, femali_list, 'ro--', label='실업율(여)', ms = 4)
plt.legend()
plt.xlabel('년도')    # x축 라벨
plt.title('출산율 / 결혼적령기 인구 실업율')
plt.xticks(year)

plt.show()
```

```python
# 데이터 스케일링 진행 후 시각화, 이중축 그래프 사용
fig, ax1 = plt.subplots()
ax1.plot(year, birth_list, 'mo-', label='출산율', ms = 3)
ax1.legend()

ax2 = ax1.twinx()
ax2.plot(year, age_list, 'go:', label='실업율', ms = 3)
ax2.plot(year, male_list, 'bo:', label='실업율(남)', ms = 3)
ax2.plot(year, femali_list, 'ro:', label='실업율(여)', ms = 3)
ax2.legend()
ax1.set_xlabel('년도')    # x축 라벨
plt.title('출산율 / 실업율 상관관계 비교')
plt.xticks(year)
plt.yticks()

plt.show()
```

```python
# 출산율과 실업율의 상관계수 구하기
data1 = finalDF.loc['출산율'].corr(finalDF.loc['결혼 적령기 실업율(%)'])
data2 = finalDF.loc['출산율'].corr(finalDF.loc['결혼 적령기 실업율(남자)(%)'])
data3 = finalDF.loc['출산율'].corr(finalDF.loc['결혼 적령기 실업율(여자)(%)'])
print(f'출산율 / 결혼 적령기 실업율의 상관계수는 {data1:0.4f}입니다.')
print(f'출산율 / 남성 결혼 적령기 실업율의 상관계수는 {data2:0.4f}입니다.')
print(f'출산율 / 여성 결혼 적령기 실업율의 상관계수는 {data3:0.4f}입니다.')

# 상관 계수 값의 범위는 −1부터 +1까지임, 계수의 절대값이 클수록 변수 사이에 강한 관계가 있음.
# 구해진 상관계수는 = 0.6022, 0.3432(남성), 0.7365(여성)
```
