```python
# 모듈 로딩 ------------------------------------------------------
import pandas as pd


# 파일 관련 변수 선언 --------------------------------------------
DIR_PATH='./Data/'
FILE_NAME=DIR_PATH+'품목별소비자물가지수.xlsx'
FILE_NAME1=DIR_PATH+'출산율.xlsx'
FILE_NAME2=DIR_PATH+'연간매매가격지수아파트.csv'

FILE_NAME
FILE_NAME1
FILE_NAME2
```

```python
# (1) CSV FILE => DataFrame 로딩 ------------------------------
moneyDF=pd.read_excel(FILE_NAME)
bornDF=pd.read_excel(FILE_NAME1,header=2, index_col=0)
buildDF=pd.read_csv(FILE_NAME2, encoding='cp949')

moneyDF.fillna(0, inplace=True)

moneyDF.isnull().sum()
```

# 물가지수 자료

```python
moneyDF.head()
```

```python
moneyDF1=moneyDF.drop(moneyDF.index[1:]) # 1에서 끝까지 인덱스를 삭제했습니다.
```

```python
moneyDF1.set_index(['시도별', '품목별'], inplace=True) # 시도별, 품목별을 인덱스로 지정했습니다.
```

```python
moneyDF1.drop('2021.12', axis=1, inplace=True) # 저희 조는 2011~2020년의 기준으로 분석했기에 2021년을 삭제했습니다.
```

```python
moneyDF1
```

```python
moneyDF1.rename(columns={'2010.12':'2010', '2011.12':'2011', '2012.12':'2012', '2013.12':'2013', '2014.12':'2014', '2015.12':'2015', '2016.12':'2016', '2017.12':'2017', '2018.12':'2018', '2019.12':'2019', '2020.12':'2020'}, inplace=True)
# moneyDF1.columns=bornDF1.columns.to_list() 
# 컬럼의 이름을 바꾸기 위해 rename을 사용해 바꿨습니다. 주석 처리해놓은 코드는 밑에서 추가 설명 드리겠습니다.
```

```python
moneyDF1.drop('2010', axis=1, inplace=True) # 앞서 말했듯이 2011~2020년 기준이라 2010년도를 삭제했습니다.
```

```python
moneyDF1
```

# 출산율 자료

```python
bornDF
```

```python
bornDF1=bornDF.drop(['출처:', '주석:']) # 필요한 데이터만 보기위해 출처, 주석을 삭제했습니다.
```

```python
bornDF1=bornDF1.iloc[:,41:] # 2011~2020년까지 추출을 위해 행은 다 불러왔지만 열은 41번째줄 2011년부터, 끝줄인 2020년까지 추출했습니다.
```

```python
bornDF1
```

# 부동산 지수 자료

```python
buildDF.head()
```

```python
buildDF1=buildDF.drop(buildDF.index[2:]) # 2번 인덱스부터 끝까지 삭제 했습니다.
buildDF1
```

```python
buildDF1.drop(0) # 0번도 삭제
```

```python
buildDF1=buildDF1.iloc[1:,4:]                        # iloc를 통해 
buildDF1.index=['전국 부동산지수']                   # 
buildDF1.drop('2021년', axis=1, inplace=True)        # 
buildDF1
```

```python
buildDF1.drop('2012년.1', axis=1, inplace=True)
buildDF1
```

```python
buildDF1.columns=bornDF1.columns.to_list()
```

```python
plusDF=pd.concat([moneyDF1,bornDF1,buildDF1])
plusDF
```

```python
plusDF= plusDF.astype(float)
```

```python
plusDF.iloc[0].corr(plusDF.iloc[2])
```

```python
plusDF.iloc[0].corr(plusDF.iloc[3])
```

```python
plusDF.iloc[2].corr(plusDF.iloc[3])
```

# 그래프

```python
# Matplotlib 한글 Font 설정 ------------------------
from matplotlib import font_manager as fm, rc

# 한글 폰트 설정
font_path='./Fonts/malgun.ttf'
font_name=fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)
```

```python
# 모듈 로딩 -------------------------------------------------------
import matplotlib.pyplot as plt
```

```python
# 그래프 그리기 => plot([x_data][y_data][z_data])
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.plot(plusDF.columns.to_list(), plusDF.iloc[0].to_list(), color='limegreen', label='물가')       
ax1.plot(plusDF.columns.to_list(), plusDF.iloc[3].to_list(), color='violet', label='전국 부동산 지수')    
ax1.legend()

ax2.plot(plusDF.columns.to_list(), plusDF.iloc[2].to_list(), color='dodgerblue', label='합계 출산율')
ax2.legend()

plt.title('물가율, 출산율 비교') 

ax1.set_xlabel('년도')

plt.ylabel('전국, 총지수')


plt.show()
```

```python
plusDF=pd.concat([moneyDF1,bornDF1,buildDF1])
plusDF= plusDF.astype(float)
plusDF
```
