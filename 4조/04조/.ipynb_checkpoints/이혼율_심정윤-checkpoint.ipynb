```python
# 모듈 로딩
import pandas as pd
import matplotlib.pyplot as plt

from matplotlib import font_manager as fm, rc  

# 한글 폰트 설정
font_path='./Fonts/malgun.ttf'
font_name=fm.FontProperties(fname=font_path).get_name()
rc('font',family=font_name)


# 파일 관련 변수 선언
DIR_PATH_DIV='./Data/'
FILE_NAME_DivRate=DIR_PATH_DIV+'(전국)이혼건수__조이혼율(2011~2021).xlsx'


DIR_PATH_DATA='./Data/'
FILE_NAME_Birth=DIR_PATH_DATA+'출산율.xlsx'
FILE_NAME_Marrige=DIR_PATH_DATA+'혼인건수_조혼인율(2011~2020).xlsx'

# Read File
Birth=pd.read_excel(FILE_NAME_Birth,skiprows=2,index_col=0)
Marrige=pd.read_excel(FILE_NAME_Marrige)

Divorce=pd.read_excel(FILE_NAME_DivRate)
```

## 출산율.xlsx :

```python
Birth
```

```python
# Birth(출산율) 파일 전처리

# 합계 출산율 (TFR, Total Fertility Rate) : 
        # 여성 1명이 평생동안 낳을 것으로 예상되는 평균 출생아 수를 나타낸 지표
        #연령별 출산율(ASFR)의 총합이며, 출산력 수준을 나타내는 대표적 지표
# 단위 : 천명, 가임여자 1명당 명

# 마지막 두 행 삭제
Birth.dropna(inplace=True)

# 열 인덱스 공백 '_'으로 변경
Birth.index=['출생아_수', '합계_출산율']

# 2011~2020로 제한
for i in range(1970,2011):
    Birth.drop([str(i)], axis=1,inplace=True)

Birth
```

## 혼인건수_조혼인율(2011~2020).xlsx : 

```python
# Divorce(이혼건수) 파일 전처리

# 첫 열 인덱스로 설정
Divorce.set_index(['기본항목별'],inplace=True)

# 컬럼 인덱스 2021.0 -> 2021 으로 변경
Divorce.columns=['2011',  '2012',  '2013',  '2014',  '2015',  '2016',  '2017',
        '2018',  '2019',  '2020',  '2021']
# 2021 컬럼 삭제
Divorce.drop('2021',axis=1,inplace=True)
Divorce
```

```python
# 출산율 - 이혼율 DF 합치기
Birth_Divorce=pd.concat([Birth,Divorce])
Birth_Divorce

# 상관계수 분석을 위해 행열 전환                           
Birth_Divorce=Birth_Divorce.transpose()

# 합계 출산율, 조이혼율 열 drop
Birth_Divorce.drop(['합계_출산율','조이혼율(천명당)'], axis=1, inplace=True)

# '출생아 수' 컬럼 단위 (명)으로 변경
Birth_Divorce['출생아_수']=Birth_Divorce['출생아_수']*1000

# 컬럼명 출생아_수'->'출생아_수(명)', '이혼건수(건)'->'이혼_수(건)'으로 변경
Birth_Divorce.columns=['출생아_수(명)','이혼_수(건)']

Birth_Divorce
```

```python
Birth_Divorce.plot()
```

```python
# 출생아 수 - 이혼건수 그래프 
x=Birth_Divorce.index
a=Birth_Divorce['출생아_수(명)']
b=Birth_Divorce['이혼_수(건)']


fig, ax1 = plt.subplots()
line1=ax1.plot(x, a, 'ro-', ms=4, label='출생아_수')
ax1.set_ylim(100000,600000)
ax1.set_ylabel('출생아 수')

ax2 = ax1.twinx()
line2=ax2.plot(b, 'bo-', ms=4, label='이혼_수')
ax2.set_ylim(100000, 150000)
ax2.set_ylabel('이혼 건수')

lines = line1 + line2
labels = [l.get_label() for l in lines]

ax1.set_xlabel('년도')
ax1.legend(lines, labels, loc='upper right')
```

```python
# 출생아 수 - 이혼건수 상관계수 분석
Birth_Divorce.corr()
```

#                         ===>  결혼 건수 대비 이혼율 - 출산율 비교

```python
Marrige
```

```python
# '기본항목별' 컬럼 인덱스로 설정
Marrige.set_index(['기본항목별'],inplace=True)

# '조횬인율' row drop
Marrige.drop(['조혼인율(천명당)'],inplace=True)

Marrige
```

```python
# '결혼 건수 당 이혼율' DF객체 생성
Marrige_Divorce_rate=pd.concat([Marrige,Divorce])
Marrige_Divorce_rate
```

```python
# 조이혼율 row drop
Marrige_Divorce_rate.drop('조이혼율(천명당)', inplace=True)

Marrige_Divorce_rate
```

```python
# 혼인 건수 당 이혼율 데이터 생성
DivRate=pd.DataFrame(Marrige_Divorce_rate.loc['이혼건수(건)']/Marrige_Divorce_rate.loc['혼인건수(건)'])

# DivRate 열 인덱스 
DivRate.columns=['이혼율(이혼건수/혼인건수)']

# DivRate 행, 열 전환
DivRate=DivRate.transpose()


# Marrige_Divorce_rate에 혼인 건수 당 이혼율 row 추가
Marrige_Divorce_rate=Marrige_Divorce_rate.append([DivRate])

# Marrige_Divorce_rate.append(Marrige_Divorce_rate.loc['혼인건수(건)']/Marrige_Divorce_rate.loc['이혼건수(건)'])

Marrige_Divorce_rate
```

```python
# DivRate 데이터 소수점 2자리까지
Marrige_Divorce_rate=Marrige_Divorce_rate.round(2)
Marrige_Divorce_rate
```

```python
Birth
```

```python
# 출산아 수 - 이혼율 DF 객체 생성
Birth_DivRate=pd.DataFrame([Birth.loc['출생아_수'],Marrige_Divorce_rate.loc['이혼율(이혼건수/혼인건수)']])

Birth_DivRate
```

```python
# Birth_DivRate 행, 열 전환
Birth_DivRate=Birth_DivRate.transpose()

Birth_DivRate

corr_DivRate = Birth_DivRate['이혼율(이혼건수/혼인건수)'].to_list()
```

```python
# 출산아 수 - 이혼율 상관계수 분석
Birth_DivRate.corr()
```

```python
Birth_DivRate.plot()
```

```python
# 출생아 술 - 이혼율  그래프
x=Birth_DivRate.index
a=Birth_DivRate['출생아_수']
b=Birth_DivRate['이혼율(이혼건수/혼인건수)']

fig, ax1 = plt.subplots()
line1=ax1.plot(x, a, 'ro-', ms=4, label='출생아_수')
ax1.set_ylim(100,600)
ax1.set_ylabel('(명)')

ax2 = ax1.twinx()
line2=ax2.plot(b, 'bo-', ms=4, label='이혼율')
ax2.set_ylim(0.2,0.7)
ax2.set_ylabel('이혼건수/혼인건수')

lines = line1 + line2
labels = [l.get_label() for l in lines]

ax1.set_xlabel('년도')
ax1.legend(lines, labels, loc='upper right')
```
