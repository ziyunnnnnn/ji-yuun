# 출산율과 인구밀도와의 상관관계 파악하기

- 수도권의 과도한 인구집중은 최근 주력으로 떠오르는 저출산의 요인
- 인구과잉으로 인한 생존경쟁 ===> 자발적인 출산제한 (멜더스 이론)
- 서울만 아니라 인천, 세종특별시, 경기도 모두의 인구를 더한 수도권 인구밀도와 출산율 비교

```python
# 모듈 설정
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import font_manager as fm, rc

# 한글 폰트 설정
font_path = './Fonts/malgun.ttf'
font_name = fm.FontProperties(fname=font_path).get_name()
rc('font', family=font_name)

# 경로 및 파일 지정
DIR_PATH = './Data/'
FILE_POP = DIR_PATH + 'population.csv'
FILE_BIRTH = DIR_PATH + '출산율.csv'

popDF = pd.read_csv(FILE_POP, encoding='cp949', header = 0, index_col=0).T
```

```python
popDF
```

```python
popDF = popDF[popDF['행정구역(동읍면)별'] == '계'].T
popDF.drop('행정구역(동읍면)별', inplace = True)
popDF.reset_index(inplace=True)
popDF
```

```python
for i in range(0 ,53 , 3):
    popDF = popDF.drop(i + 1)
    popDF = popDF.drop(i + 2)
popDF.set_index('행정구역(동읍면)별')
popDF.drop('2021', axis = 1, inplace=True)
popDF
```

```python
# 인덱스 값을 행정구역(동읍면)별로 다시 변환
popDF.set_index(popDF.columns[0], inplace=True)
# - 값을 NaN 값으로 변환
popDF.replace('-', np.nan, inplace=True)
# nan 값을 0으로 
popDF.fillna(0, inplace=True)
# 모든 값을 float 형으로
popDF.astype(float)
```

```python
# 출산율 데이터 프레임 객체 생성
birthDF = pd.read_csv(FILE_BIRTH, encoding='cp949', header=2, index_col = 0)
# 불필요한 행 drop
birthDF.drop(['출생아?수','출처:', '주석:'], inplace = True)
# 행 인덱스 이름 변경
birthDF.rename(index={'합계?출산율':'출산율'}, inplace=True)
# 필요한 년도만 추출(2011 ~ 2020년)
birthDF = birthDF.iloc[:, 41:]
```

```python
birthDF
```

```python
# 수도권 지역 데이터 설정
data_seoul = popDF.loc['서울특별시'].astype(float)
data_incheon = popDF.loc['인천광역시'].astype(float)
data_gyeonggi = popDF.loc['경기도'].astype(float)
data_sejong = popDF.loc['세종특별자치시'].astype(float)
    
# 수도권 데이터를 합한 데이터 프레임 설정
data_capital  = data_seoul + data_incheon + data_gyeonggi + data_sejong
data_capital = data_capital.astype(int)
capitalDF = pd.DataFrame({'수도권' : data_capital}).T
capitalDF
```

```python
 # 전국의 데이터 프레임 설정
data_all = popDF.loc['전국'].astype(float)
allDF = pd.DataFrame({'전국' : data_all}).T
allDF
```

```python
 # 필요한 연도 설정
year = []
for i in range(2011,2021):
    year.append(i)
        
# 인구밀도 계산
data_ration = []
for i in range(0, len(allDF.columns)):
    data = round(capitalDF.iloc[0][i]/allDF.iloc[0][i],3)
    data_ration.append(data)
rationDF = pd.DataFrame({'수도권 인구밀도' : data_ration}).T
rationDF.columns = allDF.columns.to_list()
result = pd.concat([capitalDF, allDF, rationDF, birthDF])
```

```python
result
```

```python
pd.options.display.float_format = '{:.3f}'.format
result
```

```python
## 그래프에 사용하기 위해 dataframe에서 list 추출
birth_list = result.loc['출산율'].to_list()
ration_list = result.loc['수도권 인구밀도'].to_list()

# 데이터 스케일링 진행 후 시각화, 이중축 그래프 사용
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(year, birth_list, 'mo-', label='출산율', ms = 4)
ax1.legend()
ax2.plot(year, ration_list, color = '#A40001', marker = 'o',label='수도권 인구밀도', ms = 4)
ax2.legend()
plt.xlabel('년도')    # x축 라벨
plt.title('출산율 / 수도권 인구밀도 상관관계 비교')
plt.xticks(year)
plt.yticks()

plt.show()
```

```python
# 데이터 스케일링 진행 후 시각화, 이중축 그래프 사용
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()
ax1.plot(year, birth_list, 'mo-', label='출산율', ms = 4)
ax1.legend()
ax2.plot(year, ration_list, color = '#A40001', marker = 'o',label='수도권 인구밀도', ms = 4)
ax2.legend()
ax2.invert_yaxis()
plt.xlabel('년도')    # x축 라벨
plt.title('출산율 / 수도권 인구밀도 상관관계 비교')
plt.xticks(year)
plt.yticks()

plt.show()
```

```python
corr = result.loc['출산율'].corr(result.loc['수도권 인구밀도'])
print(f'출산율 / 수도권 인구밀도의 상관계수는 {corr:0.3f}입니다.')
```
