```python
# 여자아이돌_wordcloud
noun_list = []
okt = Okt()
sentences_tag = []   
remove_set=['미개', '프로', '카드', '포토', '미스', '에스', '나인', '포토북', '미니','팬싸인회', '핑크', '대면', '공개 포카','특전', '미개봉', '럽셀콘', '앨범', '응원봉', '일괄', '스티커', '택포', 'DVD', '분할','공굿', 'CD', '블루레이', '소우주', '시즌','방탄소년단 소우주 블루레이', '사진','친필', '일본', '포스터','뱃지', '양도', '굿', '패키지', '포카', '사진', '풀셋', '액자','개봉', '가방', '포스터', '무대', '키링', '양도', '슬로건', '세트', '남자', '맥도날드', '구합', '포함', '니', '구성', '몽땅', '아이돌', '앨범', '초판', '디비디', '판매', '공식']
for article in f_article_title_list:
    sentences_tag.extend(okt.phrases(article))                   

    for tag in sentences_tag:
        if tag not in remove_set:
            noun_list.append(tag)

counts = Counter(noun_list)
tags = counts.most_common(50)


if platform.system() == 'Windows':
    path = r'c:\Windows\Fonts\malgun.ttf'
elif platform.system() == 'Darwin':  
    path = r'/System/Library/Fonts/AppleGothic'
else:
    path = r'/usr/share/fonts/truetype/name/NanumMyeongjo.ttf'

wc = WordCloud(font_path=path, background_color='white', width=800, height=600)

cloud = wc.generate_from_frequencies(dict(tags))
plt.figure(figsize=(15, 10))
plt.axis('off')
plt.imshow(cloud)
plt.show()
# print(dict(tags))
```
